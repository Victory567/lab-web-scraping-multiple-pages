```python
# Scraping other list of hot songs from last.fm
import requests
r = requests.get('http://last.fm/music/500+Greatest+Rock+&+Roll+Songs/+tracks?page=1')
r.status_code
```

```python
starts = range(1, 11) # because, 50 items per page
list(starts)
```

```python
html = r.content
fm = BeautifulSoup(html, 'html.parser')
print(fm.prettify())
```

```python
def get_song_info(fm):    
    songs = []

    for song_ in fm:
        song = song_.find('a').get_text(strip=True)
       
        songs.append(song)
    
    dct = {'songs': songs}
    
    return dct
```

```python
df = pd.DataFrame()

for start in starts:
    r = requests.get(f'https://www.last.fm/music/500+Greatest+Rock+&+Roll+Songs/+tracks?page={start}')
    soup = BeautifulSoup(r.content, 'html.parser')
    song = soup.find_all('td', attrs={'class': "chartlist-name"})
    info_dct = get_song_info(song)
    new_df = pd.DataFrame.from_dict(info_dct)
    df = pd.concat([df, new_df])

df
```

```python
df = df['songs'].str.split(' - ', expand=True)
```
